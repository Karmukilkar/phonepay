```python
import pandas as pd
import json
import os

# Path to aggregated user data
path = r"C:\Users\vinoth kanna\OneDrive\Documents\project-phonepay\pulse\data\aggregated\user\country\india\state"
Agg_state_list = os.listdir(path)

# Dictionary to hold the extracted user data
user = {'State': [], 'Year': [], 'Quater': [], 'User_brand': [], 'User_count': [], 'User_percentage': []}

# Looping through each state folder
for i in Agg_state_list:
    p_i = os.path.join(path, i)
    Agg_yr = os.listdir(p_i)

    # Looping through each year folder
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)
        Agg_yr_list = os.listdir(p_j)

        # Looping through each quarter JSON file
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            with open(p_k, 'r') as Data:
                D = json.load(Data)

                # Check if 'usersByDevice' exists and is not None
                if 'data' in D and D['data'] and 'usersByDevice' in D['data'] and D['data']['usersByDevice']:
                    for z in D['data']['usersByDevice']:
                        if all(key in z for key in ['brand', 'count', 'percentage']):
                            user['User_brand'].append(z['brand'])
                            user['User_count'].append(z['count'])
                            user['User_percentage'].append(z['percentage'])
                            user['State'].append(i)
                            user['Year'].append(j)
                            user['Quater'].append(int(k.strip('.json')))

# Create the DataFrame
Aggregated_User = pd.DataFrame(user)
Aggregated_User["State"] = Aggregated_User["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Aggregated_User["State"] = Aggregated_User["State"].str.replace("-"," ")
Aggregated_User["State"] = Aggregated_User["State"].str.title()
Aggregated_User['State'] = Aggregated_User['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")
```

```python
Aggregated_User['Year'] = Aggregated_User['Year'].astype(int)
Aggregated_User["User_count"] = Aggregated_User['User_count'].astype(int)
Aggregated_User['User_percentage'] = Aggregated_User['User_percentage'].astype(float)
Aggregated_User['Quater'] = Aggregated_User['Quater'].astype(int)
Aggregated_User.info()
Aggregated_User.to_csv(" Aggercated_User.csv", index=False)
```

```python
import pandas as pd
import json
import os

# Path to aggregated insurance data
path = r"C:\Users\vinoth kanna\OneDrive\Documents\project-phonepay\pulse\data\aggregated\insurance\country\india\state"
Agg_state_list = os.listdir(path)

# Dictionary to store the extracted data
insurance = {
    'State': [], 'Year': [], 'Quater': [],
    'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []
}

# Loop through state folders
for i in Agg_state_list:
    p_i = os.path.join(path, i)
    Agg_yr = os.listdir(p_i)

    # Loop through year folders
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)
        Agg_yr_list = os.listdir(p_j)

        # Loop through quarter JSON files
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            with open(p_k, 'r') as Data:
                D = json.load(Data)

                if 'data' in D and D['data']:
                    data_section = D['data']
                    if 'transactionData' in data_section and data_section['transactionData']:
                        for z in data_section['transactionData']:
                            if (
                                'name' in z and
                                'paymentInstruments' in z and
                                isinstance(z['paymentInstruments'], list) and
                                len(z['paymentInstruments']) > 0 and
                                'count' in z['paymentInstruments'][0] and
                                'amount' in z['paymentInstruments'][0]
                            ):
                                insurance['Transaction_type'].append(z['name'])
                                insurance['Transaction_count'].append(z['paymentInstruments'][0]['count'])
                                insurance['Transaction_amount'].append(z['paymentInstruments'][0]['amount'])
                                insurance['State'].append(i)
                                insurance['Year'].append(j)
                                insurance['Quater'].append(int(k.strip('.json')))

# Create the DataFrame
Aggregated_Insurance = pd.DataFrame(insurance)
Aggregated_Insurance["State"] = Aggregated_Insurance["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Aggregated_Insurance["State"] = Aggregated_Insurance["State"].str.replace("-"," ")
Aggregated_Insurance["State"] = Aggregated_Insurance["State"].str.title()
Aggregated_Insurance['State'] = Aggregated_Insurance['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

# Display the DataFrame
Aggregated_Insurance.info()
Aggregated_Insurance.tail()
```

```python
Aggregated_Insurance['Year'] = Aggregated_Insurance['Year'].astype(int)
```

```python
import pandas as pd
import json
import os

# Path to map user data
path = r"C:\Users\vinoth kanna\OneDrive\Documents\project-phonepay\pulse\data\map\user\hover\country\india\state"
Map_state_list = os.listdir(path)

# Dictionary to store simplified map user data
map_user = {
    'State': [], 'Year': [], 'Quater': [],
    'District': [], 'Registered_users': []
}

# Loop through states
for i in Map_state_list:
    p_i = os.path.join(path, i)
    Map_yr = os.listdir(p_i)

    # Loop through years
    for j in Map_yr:
        p_j = os.path.join(p_i, j)
        Map_quarters = os.listdir(p_j)

        # Loop through quarters
        for k in Map_quarters:
            p_k = os.path.join(p_j, k)
            with open(p_k, 'r') as Data:
                D = json.load(Data)

                if 'data' in D and D['data']:
                    hover_data = D['data'].get('hoverData', {})

                    for district, values in hover_data.items():
                        if 'registeredUsers' in values:
                            map_user['State'].append(i)
                            map_user['Year'].append(j)
                            map_user['Quater'].append(int(k.strip('.json')))
                            map_user['District'].append(district)
                            map_user['Registered_users'].append(values['registeredUsers'])

# Create the DataFrame
Map_User = pd.DataFrame(map_user)

Map_User["State"] = Map_User["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_User["State"] = Map_User["State"].str.replace("-"," ")
Map_User["State"] = Map_User["State"].str.title()
Map_User['State'] = Map_User['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

# Display the first few rows
#print(Map_User.head())
Map_User.info()
Map_User.head(24)
```

```python
import pandas as pd
import json
import os

# Path to the map_map data
path = r"C:\Users\vinoth kanna\OneDrive\Documents\project-phonepay\pulse\data\map\transaction\hover\country\india\state"
map_states = os.listdir(path)

map_map = {
    'State': [], 'Year': [], 'Quater': [],
    'District': [], 'Transaction_count': [], 'Transaction_amount': []
}

for state in map_states:
    state_path = os.path.join(path, state)
    years = os.listdir(state_path)

    for year in years:
        year_path = os.path.join(state_path, year)
        quarters = os.listdir(year_path)

        for quarter in quarters:
            file_path = os.path.join(year_path, quarter)
            with open(file_path, 'r') as f:
                data = json.load(f)

                if 'data' in data and 'hoverDataList' in data['data']:
                    for item in data['data']['hoverDataList']:
                        name = item['name']
                        metric = item.get('metric', [{}])[0]  # first item in the list
                        count = metric.get('count')
                        amount = metric.get('amount')

                        if count is not None and amount is not None:
                            map_map['State'].append(state)
                            map_map['Year'].append(year)
                            map_map['Quater'].append(int(quarter.strip('.json')))
                            map_map['District'].append(name)
                            map_map['Transaction_count'].append(count)
                            map_map['Transaction_amount'].append(amount)

# Create the DataFrame
Map_Map = pd.DataFrame(map_map)

Map_Map["State"] = Map_Map["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_Map["State"] = Map_Map["State"].str.replace("-"," ")
Map_Map["State"] = Map_Map["State"].str.title()
Map_Map['State'] = Map_Map['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

# Preview the data
Map_Map.info()
Map_Map.head(2)
```

```python
import pandas as pd
import json
import os

path = r"C:\Users\vinoth kanna\OneDrive\Documents\project-phonepay\pulse\data\map\insurance\hover\country\india\state"
states = os.listdir(path)


map_insurance = {
    'State': [], 'Year': [], 'Quater': [],
    'District': [], 'Insurance_count': [], 'Insurance_amount': []
}

for state in states:
    state_path = os.path.join(path, state)
    years = os.listdir(state_path)

    for year in years:
        year_path = os.path.join(state_path, year)
        quarters = os.listdir(year_path)

        for quarter in quarters:
            file_path = os.path.join(year_path, quarter)
            with open(file_path, 'r') as f:
                data = json.load(f)

                # Navigate and extract only TOTAL insurance metrics
                hover_data = data.get("data", {}).get("hoverDataList", [])
                for region in hover_data:
                    name = region["name"]
                    metric = region.get("metric", [])
                    for entry in metric:
                        if entry.get("type") == "TOTAL":
                            map_insurance["State"].append(state)
                            map_insurance["Year"].append(year)
                            map_insurance["Quater"].append(int(quarter.strip(".json")))
                            map_insurance["District"].append(name)
                            map_insurance["Insurance_count"].append(entry.get("count"))
                            map_insurance["Insurance_amount"].append(entry.get("amount"))

# Convert to DataFrame
Map_Insurance = pd.DataFrame(map_insurance)


Map_Insurance['State'] = Map_Insurance["State"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
Map_Insurance["State"] = Map_Insurance["State"].str.replace("-"," ")
Map_Insurance["State"] = Map_Insurance["State"].str.title()
Map_Insurance['State'] = Map_Insurance['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")


# View a sample
Map_Insurance.info()
```
