TOP INSURANCE

```python
ins=pd.read_csv("Top_Insurance.csv")
```

```python
#Null values in the data
ins.isna().sum()
#All the missing pincodes are in the same state
#Based on this we can go with median based impuatation
ins[ins["Pincode"].isna()]
```

```python
ins.fillna(ins[ins["State"]=="ladakh"]["Pincode"].median(), inplace=True)
```

```python
cursor.execute("""Create table Top_insurance(State varchar(70),
               Year YEAR,
               Quarter int,
               Pincode int,
               Transaction_count BIGINT,
               Transaction_amount decimal(18,5))""")
```

```python
ins.dropna(inplace=True)
for i,row in ins.iterrows():
    cursor.execute("""
    INSERT INTO Top_insurance(State, Year, Quarter, Pincode, Transaction_count, Transaction_amount) VALUES(%s, %s, %s, %s, %s, %s)""",
                   (row['State'], row['Year'], row['Quarter'], row['Pincode'], row['Transaction_count'], row['Transaction_amount'])
    )
    connection.commit()
print("Record inserted successfully")        
```
