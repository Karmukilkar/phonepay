## Exploratary Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import plotly.express as po
```

```python
df=pd.read_csv(" Aggercated_User.csv")
df.head()
df.shape
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x="User_brand", y="User_count", data=df)
plt.show()
```

```python
df1=pd.read_csv("Aggregated_Insurance.csv")
df1.head()
```

```python

yearly = df1.groupby("Year")["Transaction_amount"].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(yearly["Year"], yearly["Transaction_amount"], marker='o')
plt.xlabel("Year")
plt.ylabel("Total Transaction Amount")
plt.title("Yearly Transaction Amount Trend")
plt.grid(True)
```

```python

state_totals = df1.groupby('State')['Transaction_amount'].sum()
state_totals = state_totals.sort_values(ascending=False)
top8 = state_totals.head(8)
plt.figure(figsize=(6,8))
plt.pie(top8, labels=top8.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 8 States by Transaction Amount')
plt.show()
```

```python
df2=pd.read_csv("Map_Map.csv")
df2.head()
```

```python

district_totals = df2.groupby("District")["Transaction_amount"].sum().reset_index()
district_totals = district_totals.sort_values(by="Transaction_amount", ascending=False)
top8 = district_totals.head(8)

plt.figure(figsize=(5,5))
plt.pie(top8["Transaction_amount"], labels=top8["District"], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Districts by Transaction Amount')
plt.show()
```

```python
df3=pd.read_csv("Map_Insurance.csv")
df3.head()
```

```python
sorte=df3.groupby("State")["Insurance_amount"].sum().reset_index()
sorte = sorte.sort_values(by="Insurance_amount", ascending=False)
top8 = sorte.head(8)
plt.figure(figsize=(10,5))
plt.barh=sns.barplot(x="State", y="Insurance_amount", data=top8)
plt.title('Top 8 States by Insurance Amount')   
plt.xlabel('Insurance Amount')
plt.ylabel('State')
plt.show()
```
